```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup


options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/home/sandeepa/Downloads/chromedriver", options=options)
```

```python

def fetchData(query):
    driver.get("https://www.flipkart.com/")
    button=driver.find_element_by_class_name("_3Njdz7")
    if button.is_displayed():
        driver.execute_script("arguments[0].click();", button)
    
    elem = driver.find_element_by_name("q")
    elem.clear()
    elem.send_keys(query)
    elem.send_keys(Keys.RETURN)
```

```python
def clickButton(prodURL):
    driver.get(prodURL)
    try:
        button=driver.find_element_by_class_name("_2lY4Z6")
        if button.is_displayed():
            driver.execute_script("arguments[0].click();", button)
            time.sleep(1)
        return driver.page_source
    except:
        return driver.page_source

def getOffers(prodURL):
    prodPage=clickButton(prodURL)
    prodSoup=BeautifulSoup(prodPage,'html.parser')
    offers=prodSoup.find_all('span',class_="_7g_MLT row")
    return offers
```

```python
def get_products(soup):
    products=soup.find_all('div',class_="_3liAhj")
    i=1
    for product in products:

        title=product.find('a',class_="_2cLu-l")
        print(title.text)
        link='https://www.flipkart.com'+title.get('href')

        price=product.find('div',class_="_1vC4OE")
        print(price.text)

        ratings=product.find('div',class_="hGSR34")
        if(ratings==None):
            print('"No ratings for the product"')
        else:
            print(ratings.text)

        offers=getOffers(link)
        for offer in offers:
            off=offer.find('li',class_="_2-n-Lg col")
            print(off.text)
        print('-'*100)
        if(i==1):
            break
        i=i+1
  
```

```python
def get_next_page(soup):
    next_page=soup.findAll('a',class_="_3fVaIS")
    next_page_link='https://www.flipkart.com'+(next_page[len(next_page)-1].get('href'))
    return next_page_link
```

```python
query='realme'
fetchData(query)

count=0
url=driver.current_url
main_url=url
while True:
    if count!=0:
        page=requests.get(main_url)    
        soup = BeautifulSoup(page.text, 'html.parser')
        new_url=get_next_page(soup)
        url=new_url
        main_url=new_url
    
    print(url)
    page=requests.get(url)      
    soup = BeautifulSoup(page.text, 'html.parser')
    get_products(soup)
        
    if(count==10):
        break
    count=count+1
    
```
